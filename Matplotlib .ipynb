


# 2/ Chart (Labeled Pie)

```python
from matplotlib import pyplot as plt

# the data just below is used as an example 
# the lists are iterated in the same order (limits = 1 = 3%, derivatives = 3 = 10%, integrals = 10 = 33%, diff eq = 15 = 50%, applications = 1 = 3%)
# it turns all like a cycle from right to the left. If you change the order you will see it on the graph 
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications'] # is a list
num_hardest_reported = [1, 3, 10, 15, 1] # is a list



# Creation plot here
plt.figure(figsize=(10,8))
plt.pie(num_hardest_reported, labels=unit_topics, autopct="%1d%%") # "num_hardest_reported" are the slices and labeled with the corresponding topics. "autopct" puts a percentage label on each slice 

plt.axis('equal') # Set axes to be 'equal'
plt.title('Hardest Topics')

plt.show()
plt.savefig("my_pie_chart.png")
```



# 3/ Shaded Error Line

```python
from matplotlib import pyplot as plt

# the data here below are used for the example
hours_reported =[3, 2.5, 2.75, 2.5, 2.75, 3.0, 3.5, 3.25, 3.25,  3.5, 3.5, 3.75, 3.75,4, 4.0, 3.75,  4.0, 4.25, 4.25, 4.5, 4.5, 5.0, 5.25, 5, 5.25, 5.5, 5.5, 5.75, 5.25, 4.75]
exam_scores = [52.53, 59.05, 61.15, 61.72, 62.58, 62.98, 64.99, 67.63, 68.52, 70.29, 71.33, 72.15, 72.67, 73.85, 74.44, 75.62, 76.81, 77.82, 78.16, 78.94, 79.08, 80.31, 80.77, 81.37, 85.13, 85.38, 89.34, 90.75, 97.24, 98.31]



# plot part below
hours_lower_bound = [element - (element * 0.2) for element in hours_reported] # Create your hours_lower_bound  which has the elements of hours_reported minus 20% (is a list comprehension)
hours_upper_bound = [element + (element * 0.2) for element in hours_reported] # Create your hours_upper_bound which has the elements of hours_reported plus 20% (is a list comprehension)

plt.figure(figsize=(10,8))
plt.plot(exam_scores, hours_reported, linewidth=2) # "exam_scores" on the x-axis and the "hours_reported" on the y-axis, with a linewidth of 2

plt.fill_between(exam_scores, hours_lower_bound, hours_upper_bound, alpha = 0.1) # alpha is the "shading" factor/intensity .We need to precise the lower and upper bound for y-axis

plt.title("Time spent studying vs final exam scores")
plt.xlabel("Score-examen")
plt.ylabel('Hours studying')

plt.show()

plt.savefig("my_line_graph.png")
```



# 4/ Stacked Bars ("layer-on-layer")

```python
from matplotlib import pyplot as plt
import numpy as np

# data below is used as example (mathematics chapters and their number of grades obtained)
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]

x = range(5)



#create plot here

#create c_bottom, d_bottom and f_bottom here by using the Numpy library (via "np.add" function)
c_bottom = np.add(As, Bs) # Cs bars will go on top of the As + Bs = c_bottom
d_bottom = np.add(c_bottom, Cs) # Ds will be at the height of the As + Bs + Cs = c_bottom + Cs, so Cs + c_bottom (its basis) will be d_bottom (red basis)
f_bottom = np.add(d_bottom, Ds)  # Fs will be at the height of the As + Bs + Cs + Ds + Fs= d_bottom + Ds

plt.figure(figsize=(10,8))

plt.bar(x, As) # first things to be plot are As (number of A get for the exam)
plt.bar(x, Bs, bottom=As) # Bs goes on As (so bottom of Bs is As)
plt.bar(x, Cs, bottom=c_bottom) #Cs goes on Bs (so bottom of Cs is Bs)
plt.bar(x, Ds, bottom=d_bottom) #Ds goes on Cs (so bottom of Ds is Cs)
plt.bar(x, Fs, bottom=f_bottom) #Fs goes on Ds (so bottom of Fs is Ds)

ax = plt.subplot()
ax.set_xticks(range(len(unit_topics))) # creates positions on x-axis needed for the labeling  
ax.set_xticklabels(unit_topics) # attributes names of the corresponding topic to the bars

plt.title('Grade distribution')
plt.xlabel('Unit')
plt.ylabel('Number of Students')
plt.show()

plt.savefig('my_stacked_bar.png')
```


# 5/ Side-By-Side Bars

```python
from matplotlib import pyplot as plt

#these data are used for the example
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
middle_school_a = [80, 85, 84, 83, 86] 
middle_school_b = [73, 78, 77, 82, 86]



# plot bars below

# t = Number of datasets (how many different types do we have?   (here 2))
# n = number of the dataset (out of t,"which one are we talking about?")
# w = Width of each bar
# d = number of sets of bars     (here 5 sets visualized)
def create_x(t, w, n, d):
    return [t*x + w*n for x in range(d)] # this list comprehension needs always to be used for the Side-By-Side Bars


school_a_x = create_x(2, 0.8, 1, 5) # school_a_x = [0.8, 2.8, 4.8, 6.8, 8.8] -> x-axis for blue bars (school A)
school_b_x = create_x(2, 0.8, 2, 5) # school_b_x = [1.6, 3.6, 5.6, 7.6, 9.6] -> x-axis for orange bars (school B)


plt.figure(figsize=(10,8)) # creates a figure of width 10 and height 8

ax = plt.subplot() # creates a set of axes and save them to ax

# plots a set of bars (2 are ploted separetely but their positions made them appear side-by-side in-fine)
plt.bar(school_a_x, middle_school_a)
plt.bar(school_b_x,middle_school_b)

# remember zip : puts same items-order (position in list) from different lists together in  a tuple
middle_x = [ (a + b) / 2.0 for a, b in zip(school_a_x, school_b_x)] # middle_x = [1.2000000000000002, 3.2, 5.199999999999999, 7.199999999999999, 9.2]

ax.set_xticks(middle_x) # creates x-axis position for the labeling (in middle of the orange and blue bar)
ax.set_xticklabels(unit_topics) # labeling x-axis with the names from the list

plt.legend(["Middle School A", "Middle School B"])
plt.title('Test Averages on Different Units')
plt.ylabel('test average')
plt.xlabel('unit')

plt.show()

plt.savefig('my_side_by_side.png')
```


# 6/ Error Bar Chart

```python
from matplotlib import pyplot as plt

# data used fot the illustration here
past_years_averages = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
error = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4] # for error bars




# plot part here
plt.figure(figsize=(10,8)) # creates a figure of width 10 and height 8

plt.bar(range(len(past_years_averages)), past_years_averages, yerr = error, capsize=8) # yerr = y-axis with the error bars, capsize = size of the bars

plt.axis([-0.5, 6.5, 70, 95]) # sets the axis to go from -0.5 to 6.5 on the x-axis and 70 to 95 on the y-axis

ax = plt.subplot() # Creates an ax object
ax.set_xticks(range(len(years))) # x-axis ticks to be range(len(years)) for the labeling below
ax.set_xticklabels(years)  # put the names on prepared x-axis

plt.title('Final Exam Averages')
plt.ylabel('Test average')
plt.xlabel('Year')

plt.savefig('my_bar_chart.png')

plt.show()
```


# 8/ Plot with different sub-plots

```python
from matplotlib import pyplot as plt

# data used for the example
x = range(7)
straight_line = [0, 1, 2, 3, 4, 5, 6]
parabola = [0, 1, 4, 9, 16, 25, 36]
cubic = [0, 1, 8, 27, 64, 125, 216]




# making plots here

# we always spezifie the location of a plot and then we plot it afterwards
# "subplot" is based on the fact that we have a grid with 2 rows and 2 columns max, with plots taking +/- place on this space-grid
plt.subplot(2, 1, 1) # location of the first plot takes 2 rows and 1 columns in itself, and is the first element (takes the place of 2 plots)
plt.plot(x, straight_line) # and we plot it on the grid-space!

plt.subplot(2, 2, 3) # plot takes place in a grid of 2 rows, with 2 columns and is the 3rd (because plit above take the place of 2) 
plt.plot(x, parabola) # and plot it!

plt.subplot(2, 2, 4) # plot takes place in a grid of 2 rows, with 2 columns and is the 4th
plt.plot(x, cubic)

plt.subplots_adjust(wspace = 0.20, bottom =0.2) # bottom = space to the bottom of graph by changing bottom margin to 0.2 (instead default of 0.1)

plt.show()


'''
instead of the "wspace", we could also use:

- left :    left-side margin, with a default of 0.125.  increase this number to make room for a y-axis label
- right :   right-side margin, with a default of 0.9.  increase this to make more room for the figure, or decrease it to make room for a legend
- bottom :  bottom margin, with a default of 0.1. increase this to make room for tick mark labels or an x-axis label
- top :     top margin, with a default of 0.9
- wspace :  horizontal space between adjacent subplots, with default of 0.2
- hspace :  vertical space between adjacent subplots, with default of 0.2

'''

```
