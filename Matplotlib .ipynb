# This is a tutorial to learn how to use Matplotlib in order to plot different kind of graphs with your data.

# Plot of 2 histograms

```python
import matplotlib as mpl
from matplotlib import pyplot as plt

# 2 exam scores samples that we want compare visualy 
exam_scores1 = [62.55, 67.66, 81.37, 52.54, 62.98, 72.15, 59.05, 73.85, 97.24, 76.81, 89.34, 74.44, 68.52, 85.13, 90.75, 70.29, 75.62, 85.38, 77.82, 98.31, 79.08, 61.72, 71.33, 80.77, 80.31, 78.16, 61.12, 64.98, 72.66, 78.93]
exam_scores2 = [72.39, 71.27, 79.24, 83.84, 84.42, 79.38, 75.51, 76.63, 81.48,78.81,79.23,74.38,79.27,81.07,75.42,90.35,82.93,86.74,81.33,95.1,86.57,83.66,85.58,81.87,92.14,72.15,91.64,74.21,89.04,76.54,81.9,96.5,80.05,74.77,72.26,73.23,93.8,66.23,70.10,77.21]

# ploting part below
plt.figure(figsize=(10, 8)) # Create a figure of width 10 and height 8.
plt.hist(exam_scores1, bins=12, normed=True, histtype='step', linewidth=2) # Make histogram of exam_scores1, normalized, with 12 bins (histtype = 'step' makes histograms more visible)
plt.hist(exam_scores2, bins=12, normed=True, histtype='step', linewidth=2) # Make histogram of exam_scores2, normalized, with 12 bins (histtype = 'step' makes histograms more visible)

plt.legend(["1st Yr Teaching", "2nd Yr Teaching"])

plt.title("Final Exam Score Distribution")
plt.xlabel("Percentage")
plt.ylabel("Frequency")

plt.show()

plt.savefig('my_histogram.png')
```
